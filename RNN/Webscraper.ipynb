```python
import os
import csv
import newspaper
from newspaper import Article
from newspaper import fulltext
from newspaper import Source

sites = ['https://www.cnn.com', 'https://abcnews.go.com', 'https://abcnews.go.com', 'https://www.nytimes.com', 
        'https://news.yahoo.com', 'http://www.huffingtonpost.com', 'http://www.foxnews.com', 'http://www.nbcnews.com',
        'http://www.washingtonpost.com', 'http://www.theguardian.com', 'http://www.wsj.com', 'http://www.usatoday.com',
        'http://www.latimes.com']

class WebScrape:
    def __init__(self):
        self.sites = []

    def build(self, urls): # list of urls
        for url in urls:
            website = newspaper.build(url, memoize_articles=False)
            self.sites.append(website)
    
    def output(self): # output to csv files
        for website in self.sites:
            if not os.path.isdir('/Users/weichiang/Desktop/Programming/Specious/Speciousness-/news/'+ website.brand):
                os.mkdir('/Users/weichiang/Desktop/Programming/Specious/Speciousness-/news/'+ website.brand)
            
            for article in website.articles[:1]:
                article.download()
                article.parse()
                title = article.title
                # date = article.publish_date[18:22]
                main_text = article.text

                name = title + ".csv"
                file_name = "_".join(name.split())
                print(file_name)
                if not os.path.exists(os.getcwd() + "/" + file_name):
                    f = open(file_name, "w+")
                    with open(file_name, 'w') as writeFile:
                        writer = csv.writer(writeFile)
                        writer.writerows([['Title', 'Text'], [title, main_text]])
                    writeFile.close()

web = WebScrape()
web.build(['https://www.cnn.com'])
web.output()
# for article in cnn_paper.articles[:20]:
#     article.download()
#     article.parse()
#     print(article.text)
# for article in cnn_news.articles:
#     article.download()
#print(cnn_news.articles)

# cnn_paper = Source('http://cnn.com', memoize_articles=False)
# cnn_paper.build()
# print(cnn_paper.size())
```
